# Teorema Central de Límite
En este código haremos una visualización de la convergencia del teorema central del límite

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

**Pseudo-Código para visualizar la convergecia del TCL**

Sea $N \in \mathbb{N}$ el número de iteraciones. Supongamos que tenemos una colección de variables aleatorias $(X_n)_{n=1}^N$ independientes e idénticamente distribuidas con media $\mu$ y varianza $\sigma^2$, ambas finitas.

1. Para cada $m \in \{1,2,...,N\}$, calculamos $\displaystyle \bar{X}_m = \frac{\sum_{i=1}^mX_i}{m}$ y $\sigma_m = \frac{\sigma}{\sqrt{m}}$ y con ello $S_m = \frac{\bar{X}_m - \mu}{\sigma_m}$.
3. Visualizamos el histograma de la muestra $S_1, S_2, ...S_N$

Ahora haremos este proceso para diferentes distribuciones. A continuación escribiremos una función para hacer este proceso fácilmente.


```python
def TCL_fun(muestra, mu, sigmacuadrada, N, graficar = True):
  tcl = np.zeros(N) # Vector para la muestra
# El proceso de medias normalizadas
  for i in range(N):
    n_i = i + 1
    suma = np.mean(muestra[:n_i])  # X barra
    sigma_sum = math.sqrt(math.sqrt(sigmacuadrada)/n_i) # sigma_m
    tcl[i] = (suma - mu) / sigma_sum
  if graficar == True:
    plt.figure()
    plt.hist(tcl, bins=30, density=True, color = 'indigo')
    plt.title('Distribución de medias')
    plt.xlabel('Valor')
    plt.show()
  return tcl
```
